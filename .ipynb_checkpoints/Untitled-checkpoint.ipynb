```python
import math as m
import pandas as pd
```

```python
def f(x):
    y= m.e**x  - x**3
    return y
def f1(x):
    y = m.e**x - 3*(x**2)
    return y
def sign(x):
    if x < 0:
        return False
    return True
```

```python
a,b=2,5
t=0
for i in range(10):
    A=f(a)
    B=f(b)
    
    if A*B<0:
        t= (a+b)/2
        T=f(t)
        print(f'Iter = {i},a = {a},b = {b},t = {t},A = {A},B = {B},T = {T}')
        if sign(T) == sign(A):
            a = t
        else : b = t
    
```
