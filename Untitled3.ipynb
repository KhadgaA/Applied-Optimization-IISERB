```python
from ortools.sat.python import cp_model


def main():
    # Data.
    num_subjects = 6
    num_classes = 5
    num_days = 5
    all_subjects = range(num_subjects)
    all_classes = range(num_classes)
    all_days = range(num_days)

    # Creates the model.
    model = cp_model.CpModel()

    # Creates class variables.
    # classes[(n, d, s)]: subject 'n' takes class 's' on day 'd'.
    classes = {}
    for n in all_subjects:
        for d in all_days:
            for s in all_classes:
                classes[(n, d,
                        s)] = model.NewBoolVar('class_n%id%is%i' % (n, d, s))

    # Each class is assigned to exactly one subject in the schedule period.
    for d in all_days:
        for s in all_classes:
            model.Add(sum(classes[(n, d, s)] for n in all_subjects) == 1)

    # Each subject takes at most one class per day.
    for n in all_subjects:
        for d in all_days:
            model.Add(sum(classes[(n, d, s)] for s in all_classes) <= 1)

    # Try to distribute the classes evenly, so that each subject takes
    # min_classes_per_subject classes. If this is not possible, because the total
    # number of classes is not divisible by the number of subjects, some subjects will
    # be assigned one more class.
    min_classes_per_subject = (num_classes * num_days) // num_subjects
    if num_classes * num_days % num_subjects == 0:
        max_classes_per_subject = min_classes_per_subject
    else:
        max_classes_per_subject = min_classes_per_subject + 1
    for n in all_subjects:
        num_classes_takeed = []
        for d in all_days:
            for s in all_classes:
                num_classes_takeed.append(classes[(n, d, s)])
        model.Add(min_classes_per_subject <= sum(num_classes_takeed))
        model.Add(sum(num_classes_takeed) <= max_classes_per_subject)

    # Creates the solver and solve.
    solver = cp_model.CpSolver()
    solver.parameters.linearization_level = 0
    # Enumerate all solutions.
    solver.parameters.enumerate_all_solutions = True


    class subjectsPartialSolutionPrinter(cp_model.CpSolverSolutionCallback):
        """Print intermediate solutions."""

        def __init__(self, classes, num_subjects, num_days, num_classes, limit):
            cp_model.CpSolverSolutionCallback.__init__(self)
            self._classes = classes
            self._num_subjects = num_subjects
            self._num_days = num_days
            self._num_classes = num_classes
            self._solution_count = 0
            self._solution_limit = limit

        def on_solution_callback(self):
            self._solution_count += 1
            print('Solution %i' % self._solution_count)
            for d in range(self._num_days):
                print('Day %i' % d)
                for n in range(self._num_subjects):
                    is_takeing = False
                    for s in range(self._num_classes):
                        if self.Value(self._classes[(n, d, s)]):
                            is_takeing = True
                            print('  subject %i takes class %i' % (n, s))
                    if not is_takeing:
                        print('  subject {} has no class'.format(n))
            if self._solution_count >= self._solution_limit:
                print('Stop search after %i solutions' % self._solution_limit)
                self.StopSearch()

        def solution_count(self):
            return self._solution_count

    # Display the first five solutions.
    solution_limit = 5
    solution_printer = subjectsPartialSolutionPrinter(classes, num_subjects,
                                                    num_days, num_classes,
                                                    solution_limit)

    solver.Solve(model, solution_printer)

#     # Statistics.
#     print('\nStatistics')
#     print('  - conflicts      : %i' % solver.NumConflicts())
#     print('  - branches       : %i' % solver.NumBranches())
#     print('  - wall time      : %f s' % solver.WallTime())
    print('  - solutions found: %i' % solution_printer.solution_count())


if __name__ == '__main__':
    main()
```
